# import necessary packages

```python
import numpy as np
import scipy as sp
import scipy.ndimage as ni
import matplotlib.pyplot as plt

%matplotlib inline
```

# generate a binary image with two squares

```python
img = np.zeros((100, 100), dtype=np.uint8)

img[13:20, 40:45] = 1
img[67:73, 24:30] = 1

plt.imshow(img, cmap='gray', interpolation='nearest')
plt.colorbar()
```

# basic information of the image

```python
print 'type of the variable img:', type(img)
print 'data type of the variable img:', img.dtype
print 'shape of the variable img:', img.shape
```

# verify the image is binary

```python
print 'maximum value of the variable img:', np.max(img)
print 'minimum value of the variable img:', np.min(img)
```

# ndimage.label function to label separate features  

#### documentation: http://docs.scipy.org/doc/scipy-0.16.0/reference/generated/scipy.ndimage.measurements.label.html

```python
labelled, feature_num = ni.label(img)
print 'number of features:', feature_num
print 'data type of labelled image:', labelled.dtype
```

# plot result

```python
f = plt.figure(figsize=(13, 10))
plt.imshow(labelled, interpolation='nearest')
plt.colorbar()
```

# get separate binary images for each feature

```python
feature_1 = np.zeros(img.shape, dtype=np.uint8)
feature_1[labelled == 1] = 1

feature_2 = np.zeros(img.shape, dtype=np.uint8)
feature_2[labelled == 2] = 1
```

# plot result

```python
f = plt.figure(figsize = (15, 6))
ax1 = f.add_subplot(121)
ax1.imshow(feature_1, cmap='gray', interpolation='nearest')
ax1.set_title('feature 1')
ax2 = f.add_subplot(122)
ax2.imshow(feature_2, cmap='gray', interpolation='nearest')
ax2.set_title('feature 2')
```

# ndimage.binary_dilation function  

#### documentation: http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.ndimage.morphology.binary_dilation.html

```python
feature_1_dilated = ni.binary_dilation(feature_1, iterations=3, structure=[[1,1,1], [1,1,1], [1,1,1]])

f = plt.figure(figsize = (15, 6))
ax1 = f.add_subplot(121)
ax1.imshow(feature_1, cmap='gray', interpolation='nearest')
ax1.set_title('feature 1')
ax2 = f.add_subplot(122)
ax2.imshow(feature_1_dilated, cmap='gray', interpolation='nearest')
ax2.set_title('feature 1 dilated')
```

# binary dilation by using different structure

```python
feature_1_dilated_2 = ni.binary_dilation(feature_1, iterations=3, structure=[[0,1,0], [1,1,1], [0,1,0]])

f = plt.figure(figsize = (15, 6))
ax1 = f.add_subplot(121)
ax1.imshow(feature_1, cmap='gray', interpolation='nearest')
ax1.set_title('feature 1')
ax2 = f.add_subplot(122)
ax2.imshow(feature_1_dilated_2, cmap='gray', interpolation='nearest')
ax2.set_title('feature 1 dilated 2')
```

# ndimage.binary_erosion function  

#### documentation: http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.ndimage.morphology.binary_erosion.html

```python
feature_2_eroded = ni.binary_erosion(feature_2, iterations=2, structure=[[1,1,1], [1,1,1], [1,1,1]])

f = plt.figure(figsize = (15, 6))
ax1 = f.add_subplot(121)
ax1.imshow(feature_2, cmap='gray', interpolation='nearest')
ax1.set_title('feature 2')
ax2 = f.add_subplot(122)
ax2.imshow(feature_2_eroded, cmap='gray', interpolation='nearest')
ax2.set_title('feature 2 eroded')
```

# binary erosion with different structure

```python
feature_2_eroded_2 = ni.binary_erosion(feature_2, iterations=2, structure=[[0,1,0], [1,1,1], [0,1,0]])

f = plt.figure(figsize = (15, 6))
ax1 = f.add_subplot(121)
ax1.imshow(feature_2, cmap='gray', interpolation='nearest')
ax1.set_title('feature 2')
ax2 = f.add_subplot(122)
ax2.imshow(feature_2_eroded_2, cmap='gray', interpolation='nearest')
ax2.set_title('feature 2 eroded 2')
```
