# Getting Help

There are several ways of getting help in Python.

## 1) The help builtin

```python
help(open)
```

For a function or object that you have defined, "help" returns the doc-string

```python
def my_function():
    '''This is my function's doc-string.  My function doesn't do much.'''
    
```

```python
help(my_function)
```

help of course works on any object, including methods.

```python
help(hello.upper)
```

```python
hello.upper()
```

Try calling help without an argument.

```python
help()
```

You can use the builtin help facility to get information about modules or any other topic.

## 2) dir()

The command 'dir' will return all names defined in the current scope, or list all of the attributes of an object.  This is a quick way to see what methods are available for an object, like a string.  

```python
dir()
```

(For the most part, you can ignore the names returned that begin and end with an underscore; this means that they are intended to be kept "under the hood" and not used directly.  Note that one of these attributes is the doc string, intended to be accessed via 'help')

```python
print my_function.__doc__
```
