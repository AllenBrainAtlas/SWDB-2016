```python
voxel_volume = (25*25*25*((1e-6))**3)
neuron_density = 9.2*10**4/((1e-3)**3)

print voxel_volume*neuron_density
```

```python
area_volumes = {}

for area in area_acronyms:
    mask, _ = mcc.get_structure_mask(ontology[area].id)
    area_volumes[area] = np.sum(mask)*voxel_volume*neuron_density
    
print area_volumes
print np.sum(area_volumes.values())
```

```python
layer_volumes = {}

for area in area_acronyms:
    sub_areas = ontology.get_children(ontology[area].id).acronym
    sub_area_dict = {}
    for sa in sub_areas:
        mask, _ = mcc.get_structure_mask(ontology[sa].id)
        sub_area_dict[sa] = np.sum(mask)*voxel_volume*neuron_density
    sub_area_dict['total'] = np.sum(sub_area_dict.values())
    
    layer_volumes[area] = sub_area_dict
    
layer_volumes
```
