```python
drive_path = '/Volumes/Brain2016'
# We need to import these modules to get started
import numpy as np
import pandas as pd
import os
import sys
import matplotlib.pyplot as plt
%matplotlib inline
from allensdk.core.brain_observatory_cache import BrainObservatoryCache
import seaborn

manifest_path = os.path.join(drive_path,'BrainObservatory','manifest.json')
boc = BrainObservatoryCache(manifest_file=manifest_path)
expMetaData = pd.DataFrame(boc.get_ophys_experiments())
```

```python
for expContainerID in expMetaData.id[0:15]:
504593468
    exp = boc.get_ophys_experiment_data(ophys_experiment_id = expContainerID)
    moc = exp.get_motion_correction()
    xMot = abs(np.diff(moc.x_motion))
    yMot = abs(np.diff(moc.y_motion))
    mocVel = np.sqrt(xMot**2 + yMot**2)
    fig,ax = plt.subplots()
    ax.hist(mocVel,bins=np.arange(0,5,0.1),weights=np.zeros_like(mocVel) + 1. / len(mocVel))
#     plt.pause(1.0)
    ax.set_xlabel('Motion Velocity (pixel/frame)')
```

```python
expContainerID = [504593468, 504625475]
exp = boc.get_ophys_experiment_data(ophys_experiment_id = expContainerID[0])
moc = exp.get_motion_correction()
xMot = abs(np.diff(moc.x_motion))
yMot = abs(np.diff(moc.y_motion))
mocAbs = np.sqrt(moc.x_motion**2 + moc.y_motion**2)
mocVel = np.sqrt(xMot**2 + yMot**2)

stepSize = 600;

fig, ax1 = plt.subplots()
t = np.arange(len(mocAbs))/float(30)/60
s1 = mocAbs
ax1.plot(t[::stepSize], s1[::stepSize],alpha=0.6);
ax1.set_xlabel('time (min)')
# Make the y-axis label and tick labels match the line color.
# ax1.set_ylabel('exp', color='b')
# for tl in ax1.get_yticklabels():
#     tl.set_color('b')
ax1.set_ylim(0,18)


ax2 = ax1.twinx()
t = np.arange(len(mocVel))/float(30)/60
s2 = mocVel
ax2.plot(t[::stepSize], s2[::stepSize], c=(0.5,0,0),alpha=0.6);
# ax2.set_ylabel('sin', color='r')
# for tl in ax2.get_yticklabels():
#     tl.set_color('r')
plt.show()
# ax2.set_xlim(0,10)
ax2.set_ylim(0,18)

fig,ax = plt.subplots();
ax.hist(mocAbs,alpha=0.4);
ax.hist(mocVel,alpha=0.4,color=(0.5,0,0));
ax.set_xlim(0,18)
```
