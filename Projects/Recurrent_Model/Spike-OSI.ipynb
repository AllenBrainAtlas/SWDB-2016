```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def plothis(cell_type):
    mat_file = '../drifting_gratings_nonrunning/'+cell_type+cell_type+'_spikeosi.mat' #to load osi from
    matform=sio.loadmat(mat_file)
    osi=matform['osi'][0]
    plt.hist(osi, bins =50,histtype='step', range=[-1,5],log=False,normed = True,alpha=0.5,label=cell_type)
    plt.legend(loc='upper right', fontsize=15)
    return osi
```

```python
osi_L4_Cux2_VISp=plothis('L4_Cux2_VISp')
osi_L4_Rorb_VISp=plothis('L4_Rorb_VISp')
osi_L4_Scnn1a_VISp=plothis('L4_Scnn1a_VISp')
```

```python
osi_L4_all=np.concatenate((osi_L4_Cux2_VISp,osi_L4_Rorb_VISp))
osi_L4_all=np.concatenate((osi_L4_all,osi_L4_Scnn1a_VISp))
```

```python
plothis('L23_VISp')
plt.hist(osi_L4_all, bins =50,histtype='step', range=[-1,5],normed = True,log=False,alpha=0.5,label='L4_alltype')
plt.legend(loc='upper right', fontsize=15)
```
