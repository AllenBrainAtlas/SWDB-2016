```python
#Generates list of neurons and stores them into ../build/FILENAME.csv
filename = 'neuron_list1'

#volume dimensions
x_lim = [-1000, 1000]
y_lim = [-1000, 1000]
z_lim = [-50, 50]

#number of neurons
num_E = 400
num_I = 100

#neuron type
type_E = 'LIF_exc'
type_I = 'LIF_inh'
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
#generate data
col_index = range(num_E+num_I)
col_type = [type_E]*num_E + [type_I]*num_I
col_x = list(np.round(np.random.rand(num_E+num_I)*np.diff(x_lim)+x_lim[0], 3))
col_y = list(np.round(np.random.rand(num_E+num_I)*np.diff(y_lim)+y_lim[0], 3))
col_z = list(np.round(np.random.rand(num_E+num_I)*np.diff(z_lim)+z_lim[0], 3))
col_none = ['None']*(num_E+num_I)
```

```python
#save data
nrn_dict = {'index': col_index, 'type': col_type, 'x': col_x, 'y': col_y, 'z': col_z, 'tuning': col_none, 'morphology': col_none, 'cell_par': col_none}
nrn_list = pd.DataFrame(nrn_dict, columns=['index', 'type', 'x', 'y', 'z', 'tuning', 'morphology', 'cell_par'])
nrn_list.to_csv('../build/' + filename + '.csv', sep=' ', index=False)
nrn_list
```

```python
#plot positions
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(np.array(col_x)[np.array(col_type)==type_E], np.array(col_y)[np.array(col_type)==type_E], np.array(col_z)[np.array(col_type)==type_E], '.b', label='E: %d'%(num_E))
ax.plot(np.array(col_x)[np.array(col_type)==type_I], np.array(col_y)[np.array(col_type)==type_I], np.array(col_z)[np.array(col_type)==type_I], '.r', label='I: %d'%(num_I))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Positions of neurons')
ax.legend()
fig.savefig('../build/' + filename + '_pos.png', dpi=600)
```
